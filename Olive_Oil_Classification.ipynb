````python
"""
Train and test split
"""

X_train = train_df.iloc[:, :-1]
y_train = train_df["target"]


X_test = test_df.iloc[:, :-1]
y_test = test_df["target"]

metrics_df = pd.DataFrame(columns=["model", "precision", "accuracy", "recall", "f1"])

# Function helper that registers metrics to a global DF
def save_metrics(model, precision, accuracy, recall, f1, df):
  metrics = pd.Series({"model":model, "precision":precision, "accuracy":accuracy, "recall":recall, "f1": f1})
  df = pd.concat([df, metrics.to_frame().T], ignore_index=True)
  return df
````

# Naive Bayes


````python

#Import the needed model.
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn import metrics
````


 GaussianNB

````python
"""
Train the model with the training dataframe.
"""

gaussian_model = GaussianNB()
gaussian_model.fit(X_train, y_train)

y_pred = gaussian_model.predict(X_test)

# Calculate metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted')
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
conf_matrix = metrics.confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print('Confusion matrix:')
print(conf_matrix)

# Print metrics
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1_score)

metrics_df = save_metrics("Gaussian", precision, accuracy, recall, f1_score, metrics_df)
````

 MultinomialNB

````python
"""
Train the model with the training dataframe.
"""

multinomial_model = MultinomialNB()
multinomial_model.fit(X_train.abs(), y_train)

y_pred = multinomial_model.predict(X_test)

# Calculate metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted')
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
conf_matrix = metrics.confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print('Confusion matrix:')
print(conf_matrix)

# Print metrics
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1_score)

metrics_df = save_metrics("Multinomial", precision, accuracy, recall, f1_score, metrics_df)
````

# KNN

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
````

````python

"""
Training the KNN model.
"""
def knn(k):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred = knn_model.predict(X_test)

  # Calculate metrics
  accuracy = metrics.accuracy_score(y_test, y_pred)
  precision = metrics.precision_score(y_test, y_pred, average='weighted', zero_division=1)
  recall = metrics.recall_score(y_test, y_pred, average='weighted')
  f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
  conf_matrix = metrics.confusion_matrix(y_test, y_pred)

  return pd.Series({"k":k,"accuracy":accuracy,"precision":precision,"recall":recall,"f1": f1_score,"confusion_matrix": conf_matrix})

k_metrics = pd.DataFrame(columns=["k","accuracy","precision","recall","f1", "confusion_matrix"])
for k in range(1,len(X_train)):
  model_metrics = knn(k)
  k_metrics = pd.concat([k_metrics,model_metrics.to_frame().T], ignore_index=True)

k_metrics.head(25)

metrics_df = save_metrics("KNN k=3", k_metrics.iloc[2]['precision'], k_metrics.iloc[2]['accuracy'], k_metrics.iloc[2]['recall'], k_metrics.iloc[2]['f1'], metrics_df)
````

As such we notice that the best **number of neighbors** for our Dataset is 3.

````python
# Create a KNN model with the best k value
best_k_model = KNeighborsClassifier(n_neighbors=3)
best_k_model.fit(X_train, y_train)

# Make predictions on the test set using the best k model
y_pred = best_k_model.predict(X_test)

# Calculate accuracy, precision, recall, and F1 score
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted', zero_division=1)
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 score: {f1_score:.4f}")
````

# SVM

````python
from sklearn import svm, metrics
````

````python

def svm_model(kernel,metrics_df):
  svm_model = svm.SVC(kernel=kernel)
  svm_model.fit(X_train, y_train)
  # Make predictions on the test set
  y_pred = svm_model.predict(X_test)
  # Calculate metrics
  accuracy = metrics.accuracy_score(y_test, y_pred)
  precision = metrics.precision_score(y_test, y_pred, average='weighted', zero_division=1)
  recall = metrics.recall_score(y_test, y_pred, average='weighted')
  f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
  conf_matrix = metrics.confusion_matrix(y_test, y_pred)
  print('Confusion matrix:')
  print(conf_matrix)

  # Print metrics
  print('Accuracy:', accuracy)
  print('Precision:', precision)
  print('Recall:', recall)
  print('F1-score:', f1_score)
  return save_metrics("SVM " + kernel, precision, accuracy, recall, f1_score, metrics_df)

````

````python
kernels = ["linear", "poly", "rbf", "sigmoid"]
for kernel in kernels :
  metrics_df = svm_model(kernel,metrics_df)
````
